```python
%matplotlib inline
import pylab as pb
import numpy as np
from math import pi
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
import scipy.stats as st
import random
# To sample from a multivariate Gaussian
def F(mu, K):
    return np.random.multivariate_normal(mu,K);
# To compute a distance matrix between two sets of vectors
def D(x1, x2):
    return cdist(x1,x2)
# To compute the exponetial of all elements in a matrix
def E(D):
    return np.exp(D)
# To compute an error modelled by a normal distribution
def normErr(mu, cov):
    return np.random.normal(mu, cov)
# To compute the value of a point in the model
def applyModelParams(x, W, mu, cov):
    return (W[0]*x + W[1] + normErr(mu, cov))
```

```python
#prior is just an initial guess, parameters don't matter
cov0 = np.matrix([[1, 0],[0, 1]])
mu0 = np.matrix([[0],[0]])
xSpace = np.linspace(-7, 7, num=2000)
ySpace = np.linspace(-7, 7, num=2000)
xRange = np.linspace(-1,1,num=201)
XT, YT = np.meshgrid(xSpace, ySpace)
XY = np.empty(XT.shape + (2,))
XY[:, :, 0] = XT
XY[:, :, 1] = YT

def pointsPDF(mu, cov, space, grid):
    data = [F(np.ravel(mu), cov) for i in space]
    dataDim0 = [k[0] for k in data]
    dataDim1 = [k[1] for k in data]
    pdf = st.multivariate_normal.pdf(grid, np.ravel(mu), cov)
    return (dataDim0, dataDim1, pdf)

def scatterContour(XT, YT, dataDim0, dataDim1, pdf):
    fig = plt.figure(figsize=(6, 4))
    ax = fig.add_subplot( 111 )
    ax.scatter(dataDim0, dataDim1, marker='.')
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_xlabel('w_0')
    ax.set_ylabel('w_1')
    ax.set_title('The Prior Distribution Over W')
    plt.contourf(XT, YT, pdf, 50, cmap='jet')
    #plt.show()
    plt.savefig('prior.png', bbox_inches='tight')

prior = pointsPDF(mu0, cov0, xRange, XY)

scatterContour(XT, YT, prior[0], prior[1], prior[2])

def scatterContour2(XT, YT, dataDim0, dataDim1, pdf, ax):
    ax.scatter(dataDim0, dataDim1, marker='.')
    ax.contour(XT, YT, pdf, 10)


#plt.hist(yDim0, bins=200)
#plt.show()
#plt.hist(yDim1, bins=200)
#plt.show()
```

```python
#image of our model, i.e. visualising the ys, i.e. the likelihood
xArr = np.linspace(-1,1,num=201)
errorMu, errorCov = 0, 0.3
W = [-1.3, 0.5]

def likelihood(W, ax, b):
    yArr = [applyModelParams(i, W, errorMu, errorCov) for i in xArr]
    ax.plot(xArr, yArr)
    if (b):
        return yArr

def sampleLine(W, ax):
    yLine = W[0]*xArr + W[1]
    ax.plot(xArr, yLine)
        
fig = plt.figure()
ax = fig.add_subplot( 111 )
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('The Model')
yArr = likelihood(W, ax, True)
#sampleLine(W, ax)
#plt.show()
plt.savefig('model.png', bbox_inches='tight')
```

```python
#we pick a point from the data, so just draw it randomly
#find the new mean and covariance given by that point
#and find the posterior that defines.

#then repeat more times to gain more information.
#as we can see, the posterior converges onto the original parameters,
#-1.3 and 0.5
#and the lines become much closer together.

oldCov = np.matrix([[1, 0],[0, 1]])
oldMu = np.matrix([[0],[0]])
xPoints = []
yPoints = []
for i in range(50):
    #draw a random point from our model
    index = random.randint(0, len(xArr) -1)
    x, y = xArr[index], yArr[index]
    xPoints.append(x)
    yPoints.append(y)
    errorCov = 0.3
    #use this point to draw out a new mean and covariance, using the function given
    X = np.matrix([x, 1])
    newCov = np.linalg.inv(np.linalg.inv(oldCov) + 1/errorCov * X.T * X)
    newMu = newCov*(np.linalg.inv(oldCov)*oldMu + 1/errorCov*X.T*y)
    #create a new posterior function using these parameters
    posterior = [F(np.ravel(newMu), newCov) for i in xRange]
    posteriorDim0 = [k[0] for k in posterior]
    posteriorDim1 = [k[1] for k in posterior]
    posteriorPDF = st.multivariate_normal.pdf(XY, np.ravel(newMu), newCov)
    #plot this new pdf
    if (i < 5 or i == 24 or i == 49):
        newFig = plt.figure(figsize=(12, 4))
        figTitle = 'After ' + str(i+1) + ' Samples'
        newFig.suptitle(figTitle, y=1.08)
        ax0 = newFig.add_subplot ( 121 )
        ax0.scatter(posteriorDim0, posteriorDim1, marker='.', c='w')
        ax0.contourf(XT, YT, posteriorPDF, 50, cmap='jet')
        ax0.set_xlim(-3, 3)
        ax0.set_ylim(-3, 3)
        ax0.set_xlabel('w_0')
        ax0.set_ylabel('w_1')
        title0 = 'The Posterior Distribution Over W'
        ax0.title.set_text(title0)
        ax1 = newFig.add_subplot( 122 )
        ax1.plot(xArr, yArr, zorder = 1)
        ax1.set_xlim(-1, 1)
        ax1.set_ylim(-3, 3)
        ax1.set_xlabel('x')
        ax1.set_ylabel('y')
        title1 = 'Samples From The Posterior With The Points Drawn From The Model'
        ax1.title.set_text(title1)
        for j in range(8):
            W = F(np.ravel(newMu), newCov)
            yLine = W[0]*xArr + W[1]
            ax1.plot(xArr, yLine, zorder = 2)
        
        
        ax1.scatter(xPoints, yPoints, c='pink', marker='o', zorder=3) 
        figName = 'posterior-' + str(i) + '.png'
        #plt.show()
        newFig.savefig(figName, bbox_inches='tight')
    oldCov = newCov
    oldMu = newMu
```

```python
#regression with the squared exponential covariance function
#sig => sigma^{2}_{f}, the maximum allowable covariance.
#-if our function has a broad range on the y-axis, we want this number to be high
#-if xi ~= xj, squared exponential covariance should approach this number
#    (obvious when you look at the exponent I guess).
#    (also makes sense in terms of that neighbouring values should have low 
#     cov, as we want to have a smooth function)
#l => length parameter, i.e. the lengthscale
#-if two values are far apart, the effect of this separation in terms of the
# covariance is determined by l
#
#
#sqExpCov will compute the squared exponential covariance - a kernel.
#xi is a 1xN vector and xj is a 1xM vector
#We output Cov, an NxM vector

def sqExpCov(xi, xj, sig_f, l):
    xiRepeat = np.matlib.repmat(xi, xi.shape[1], 1)
    xjRepeat = np.matlib.repmat(xj, xj.shape[1], 1)    
    #diff = (D(xiRepeat, xjRepeat).T)*(D(xiRepeat, xjRepeat))
    #diff = np.dot(D(xiRepeat, xjRepeat).T, D(xiRepeat, xjRepeat))
    diff = cdist(xiRepeat, xjRepeat, 'sqeuclidean')
    Cov = (sig_f**2)*E((-diff)/(l**2))    
    return Cov

#the GP prior is defined as f(x1, ..., xn) ~ N(0, K) with K covariance matrix
#so we can use the function as defined above to be our K
#we use sigma = 1 to simplify things, as it is essentially a measure of noise
def plotGaussianProcess(l):
    X = np.linspace(-4, 4, 2000)
    X = X[:,None] #reshape X to make it n*D
    mu = np.zeros((2000))
    K = sqExpCov(X, X, 1, l)
    Z = np.random.multivariate_normal(mu, K, 20)
    gpFig = plt.figure()
    gpAx = gpFig.add_subplot( 111 )
    gpAx.set_xlim(-4, 4)
    gpAx.set_ylim(-4, 4)
    gpAx.set_xlabel('x')
    gpAx.set_ylabel('y')
    title = "Sample from the GP prior, with lengthscale = " + str(l)
    gpAx.set_title(title)
    for i in range(20):
        gpAx.plot(X[:], Z[i,:])
    #plt.show()
    gpFile = 'gpLength-' + str(l) +'.png'
    plt.savefig(gpFile, bbox_inches='tight')
    

plotGaussianProcess(20)
plotGaussianProcess(5)
plotGaussianProcess(1)
plotGaussianProcess(0.3)
plotGaussianProcess(0.0004)
#this is our prior



```

```python
def plotGp(l):
    X = np.linspace(-4.0,4.0,2000) # 500 points evenly spaced over [0,1]
    X = X[:,None] # reshape X to make it n*D
    mu = np.zeros((2000)) # vector of the means
    K =  np.exp(-cdist(X, X, 'sqeuclidean')/(l*l))# covariance matrix
    # Generate 20 sample path with mean mu and covariance C
    Z = np.random.multivariate_normal(mu,K,20)
    pb.figure() # open new plotting window
    for i in range(20):
        pb.plot(X[:],Z[i,:])
    title = 'length-scale '+str(l)
    pb.title(title)
    pb.show()

def generateData():
    X = np.array([-pi, -3*pi/4,-pi/2,0, pi/2, 3*pi/4, pi])
    epsilon = np.sqrt(0.5) * np.random.randn(7)
    Y = np.sin(X)+epsilon
    return (X,Y)

def kernel(X, Y, l=1.):
    return np.exp(-cdist(X, Y, 'sqeuclidean')/(l*l))

def computePosterior(xStar, X, Y,l):
    #Xstar = np.array([xStar])
    Xstar = xStar
    X = X[:,None]
    Xstar = Xstar[:,None]
    k = kernel(Xstar,X,l)
    C = np.linalg.inv(kernel(X,X,l))
    t = Y[:,None]
    mu = np.dot(np.dot(k,C),t)

    c = kernel(Xstar, Xstar,l)
    sigma = c- np.dot(np.dot(k,C),np.transpose(k))
    return mu, sigma

def plotSamplePos():
    X, Y = generateData()
    x = np.linspace(-2*np.pi, 2*np.pi, 800)
    mu, sigma=computePosterior(x,X,Y,2.)
    #print(mu)
    mu = np.reshape(mu,(800,))
    x = x[:,None]
    Z = np.random.multivariate_normal(mu,np.nan_to_num(sigma),20)
    pb.figure() # open new plotting window?
    pb.plot(X,Y,'ro')
    for i in range(20):
        pb.plot(x[:],Z[i,:])
    pb.show()

def plotPosterior():
    #plotGp(0.2)
    X, Y = generateData()
    x = np.linspace(-2*np.pi, 2*np.pi, 800)
    list = np.arange(0.2,2,0.2)
    print(list)
    for l in list:

        mu, sigma = computePosterior(x,X,Y,l)
        #plot observations
        plt.plot(X, Y,'ro')
        plt.plot(x,np.sin(x), color = 'green')
        mu = np.reshape(mu, (800,))
        plt.plot(x,mu, color = 'blue')
        upper = mu + 2*np.sqrt(sigma.diagonal())
        lower = mu - 2*np.sqrt(sigma.diagonal())
        ax = plt.gca()
        ax.fill_between(x, upper, lower, facecolor='orange', interpolate=True, alpha=0.2)
        #plt.title(title)
        plt.show()

plotSamplePos()
plotPosterior()









```
