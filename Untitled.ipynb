```python
import pylab as pb
import numpy as np
from math import pi
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
import scipy.stats as st
import random
# To sample from a multivariate Gaussian
def F(mu, K):
    return np.random.multivariate_normal(mu,K);
# To compute a distance matrix between two sets of vectors
def D(x1, x2):
    return cdist(x1,x2)
# To compute the exponetial of all elements in a matrix
def E(D):
    return np.exp(D)
# To compute an error modelled by a normal distribution
def normErr(mu, cov):
    return np.random.normal(mu, cov)
# To compute the value of a point in the model
def applyModelParams(x, W, mu, cov):
    return (W[0]*x + W[1] + normErr(mu, cov))
```

```python
mean = [0, 0]
cov = np.matrix([[1, 0],[0, 1]])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
#prior is just an initial guess, parameters don't matter
cov0 = np.matrix([[1, 0],[0, 1]])
mu0 = np.matrix([[0],[0]])
xSpace = np.linspace(-7, 7, num=2000)
ySpace = np.linspace(-7, 7, num=2000)
xRange = np.linspace(-1,1,num=201)
XT, YT = np.meshgrid(xSpace, ySpace)
XY = np.empty(XT.shape + (2,))
XY[:, :, 0] = XT
XY[:, :, 1] = YT

def pointsPDF(mu, cov, space, grid):
    data = [F(np.ravel(mu), cov) for i in space]
    dataDim0 = [k[0] for k in data]
    dataDim1 = [k[1] for k in data]
    pdf = st.multivariate_normal.pdf(grid, np.ravel(mu), cov)
    return (dataDim0, dataDim1, pdf)

def scatterContour(XT, YT, dataDim0, dataDim1, pdf):
    fig = plt.figure(figsize=(9, 6))
    ax = fig.add_subplot( 111 )
    ax.scatter(dataDim0, dataDim1, marker='.')
    plt.contour(XT, YT, pdf, 10)
    plt.show()

prior = pointsPDF(mu0, cov0, xRange, XY)

scatterContour(XT, YT, prior[0], prior[1], prior[2])


#plt.hist(yDim0, bins=200)
#plt.show()
#plt.hist(yDim1, bins=200)
#plt.show()
```

```python
#image of our model, i.e. visualising the ys, i.e. the likelihood
xArr = np.linspace(-1,1,num=201)
errorMu, errorCov = 0, 0.3
W = [-1.3, 0.5]

def likelihood(W, ax):
    yArr = [applyModelParams(i, W, errorMu, errorCov) for i in xArr]
    ax.plot(xArr, yArr)
    return yArr

def sampleLine(W, ax):
    yLine = W[0]*xArr + W[1]
    ax.plot(xArr, yLine)
        
fig = plt.figure()
ax = fig.add_subplot( 111 )
yArr = likelihood(W, ax)
sampleLine(W, ax)
plt.show()
```

```python
#pick a point from the data so let's just draw randomly
def drawRandomPoint(xModel, yModel):
    pointIndex = random.randint(0, len(xModel))
    x, y = xModel[pointIndex], yModel[pointIndex]
    return (x, y)

#find the new mean and covariance using what we started with and the point drawn out
def updateMeanCov(x, y, mu, cov, errorCov):
    basisVec = np.matrix([x, 1])
    newCov = np.linalg.inv(np.linalg.inv(cov) + 1/errorCov * basisVec.T * basisVec)
    newMu = cov*(np.linalg.inv(cov)*mu + 1/errorCov*basisVec.T*y)
    return(newMu, newCov)

x0, y0 = drawRandomPoint(xArr, yArr)

mu1, cov1 = updateMeanCov(x0, y0, mu0, cov0, errorCov)

#create the posterior using our updated mean and covariance
posterior = pointsPDF(mu1, cov1, xRange, XY)

scatterContour(XT, YT, posterior[0], posterior[1], posterior[2])
```

```python
#repeat a few times to gain more information
xPrev = x0
yPrev = y0
muPrev = mu1
covPrev = cov1
for i in range(5):
    x,y = drawRandomPoint(xArr, yArr)
    mu, cov = updateMeanCov(xPrev, yPrev, muPrev, covPrev, errorCov)
    newPosterior = pointsPDF(mu, cov, xRange, XY)
    
    scatterContour(XT, YT, newPosterior[0], newPosterior[1], newPosterior[2])
    
    newFig = plt.figure()
    ax = newFig.add_subplot( 111 )
    yArr = likelihood(W, ax)
    sampleLine(W, ax)
    for i in range(4):
        newW = drawRandomPoint(newPosterior[0], newPosterior[1])
        sampleLine(newW, ax)
        
    plt.show()
    
    
    xPrev = x
    yPrev = y
    muPrev = mu
    covPrev = cov
```
